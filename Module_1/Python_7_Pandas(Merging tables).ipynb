```python
FOLDERS = [
     "C:/Users/sarga/Python3/Module_1/data"
]
data1 = pd.DataFrame(columns =['UserId','MovieId','Rating','timestamp'] ) #create empty dataFrame
def concatenation_df(folder):
    global data1
    for path, dirs, files in os.walk(folder):
        for file in files:
            fileName1 = os.path.join(path,file)
            temp1 = pd.read_csv(fileName1,names = ['UserId','MovieId','Rating','timestamp'])
            data1 = pd.concat([data1,temp1])
            
for folder in FOLDERS:
    concatenation_df(folder)
print(display(data1))
```
