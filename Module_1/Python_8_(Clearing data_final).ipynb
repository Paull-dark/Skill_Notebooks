```python
# С помощью apply и лямбда-функции увеличьте возраст во всех записях на 1 год и сохраните в sample2. В переменной sample2 
# должен содержаться весь датафрейм sample.
sample2 = sample.copy()
sample2['Age'] = sample2['Age'].apply(lambda x: x+1)
sample2
```

```python
# С помощью apply и lambda-функции замените все буквы в поле City на маленькие и сохраните в sample2. 
# Вам может понадобиться функция s.lower().
# Обратите внимание: когда в столбце есть пропущенные значения, необходимо в явном виде указывать, что это str.
sample2=sample.copy()
sample2['City'] = sample2['City'].apply(lambda x: str(x).lower())
sample2
```

```python
# Примените функцию profession_code для того, чтобы заменить поле Profession с помощью apply.
# Сохраните получившийся датафрейм в переменную sample2.
# Функцию необходимо определить повторно!
def profession_code(string):
    if string == 'Рабочий':
        return 0
    if string == 'Менеджер':
        return 1
    return 2
sample2 = sample.copy()
sample2['Profession'] = sample2['Profession'].apply(profession_code)
sample2
```

```python
# Примените функцию age_category и apply, чтобы создать новую колонку в sample под названием 'Age_category'.
# Не забудьте загрузить датафрейм.

sample = pd.read_csv('sample.csv')

def age_category(age):
    if age < 23:
        return 'молодой'
    if 23<age<35:
        return 'средний'
    return 'зрелый'
    
sample['Age_category'] = sample['Age'].apply(age_category)
```

На месте записей с ошибками в user_id должна быть пустая строка "". Сделайте это через apply и новую функцию, которую вы создадите. Результат сохраните в log:

```python
log = pd.read_csv('log.csv', header=None)
log.columns = ['user_id','time','bet','win']

def upd_names(string):
    if 'user' in string:
        p1 = string.find('u')
        return string[p1:]
    return ''

log['user_id'] = log['user_id'].apply(upd_names)
```
