```python
# Отметьте позиции (Position), по которым общая сумма зарплат (Wage) игроков превышает 5 млн евро в год.

fb_po = fb.groupby(['Position'])['Wage'].sum().sort_values(ascending = False)
fb_po
```

```python
# Посчитайте среднюю (mean) и медианную (median) зарплату (Wage) футболистов из разных клубов (Club). 
# В скольких клубах средняя и медианная зарплаты совпадают?

med = fb.groupby(fb['Club'])['Wage'].agg(['mean', 'median'])
med.loc[(med['mean']) == (med['median'])].info()
```

```python
med = fb.groupby(fb['Club'])['Wage'].agg(['mean', 'median'])
med.loc[(med['mean']) == (med['median'])].sort_values(
    ['mean'], ascending=False).head(3)
```

```python
# С помощью функции groupby посчитайте сумму зарплат (Wage) футболистов клуба (Club) "Chelsea".
che = fb.groupby(fb['Club'])['Wage'].sum()
che['Chelsea']
```

```python
# Определите максимальную зарплату футболиста национальности (Nationality)
# Аргентина ("Argentina") в возрасте 20 лет.

rt = fb.loc[(fb['Age'] == 20)].groupby(fb['Nationality'])['Wage'].max()
rt['Argentina']
```

```python
# Определите максимальную зарплату футболиста национальности (Nationality) Аргентина 
#("Argentina") в возрасте 30 лет.

max_sal = fb.loc[(fb['Age'] == 30)].groupby(fb['Nationality'])['Wage'].max()
max_sal['Argentina']
```

```python
# Определите минимальную зарплату футболиста национальности (Nationality) 
# Аргентина ("Argentina") в возрасте 30 лет.

min_sal = fb.loc[(fb['Age'] == 30)].groupby(fb['Nationality'])['Wage'].min()
min_sal['Argentina']
```

```python
# Определите максимальную силу (Strength) и баланс (Balance) среди игроков клуба (Club) "FC Barcelona"
# из Аргентины  ("Argentina"). Ответ введите через точку с запятой без пробела.

tu = fb.loc[(fb['Club']) == 'FC Barcelona'].groupby(
    fb['Nationality'])[['Strength', 'Balance']].max()
tu.loc['Argentina']
```

```python
# Создайте сводную таблицу, содержащую сведения о количестве игроков, занимающих разные позиции в каждом клубе. Отсутствующие значения замените нулями.
# Каково среднее количество вратарей (GK) в клубе? Ответ округлите до трёх цифр после запятой.

piv = fb.pivot_table(values=['Name'],
                     index=['Position'],
                     columns=['Club'],
                     aggfunc='count',
                     fill_value=0,
                     )
piv.loc['GK'].mean()
```

```python
# Используя таблицу, созданную на предыдущем шаге, определите, сколько клубов не содержат данных о центральных полузащитниках. (CM)
# Подсказка: для выполнения этого задания желательно сохранить сводную таблицу в виде отдельного датафрейма и сгруппировать часть данных этого датафрейма с помощью value_counts().

piv = fb.pivot_table(values=['Name'],
                     index=['Position'],
                     columns=['Club'],
                     aggfunc='count',
                     fill_value=0,
                     )
piv.loc['CM'].value_counts().sort_values(ascending = False)
```

```python
# С помощью сводной таблицы и функции loc посчитайте, сколько получают ("Wage") 
# русские футболисты ("Russia"), играющие за ФК "AS Monaco".

piv_rus = fb.pivot_table(values = ['Wage'],
                        index = ['Nationality'],
                        columns=['Club'],
                        aggfunc = 'sum')
piv_rus.loc['Russia']['Wage']['AS Monaco']


```

```python
# Создайте сводную таблицу, содержащую сведения о средней скорости футболистов, занимающих разные позиции (Position) в разных футбольных клубах (Club)
# Основываясь на данных таблицы, отметьте три позиции, представители которых в среднем обладают самой высокой скоростью.
piv_speed = fb.pivot_table(values=['SprintSpeed'],
                          index = ['Club'],
                          columns=['Position'],
                          aggfunc='mean',
                          fill_value=0)
piv_speed.max().sort_values(ascending = False)
```

```python
# Используя таблицу, созданную на предыдущем шаге, отметьте названия трёх клубов, в которых центральные форварды (ST) обладают наибольшей средней скоростью.
piv_speed = fb.pivot_table(values=['SprintSpeed'],
                           index=['Club'],
                           columns=['Position'],
                           aggfunc='mean',
                           fill_value=0)
piv_speed.SprintSpeed['ST'].sort_values(ascending=False)
```

```python
# Используя таблицу, созданную на предыдущем шаге, отметьте названия трёх клубов, в которых центральные форварды (ST) обладают наибольшей средней скоростью.
piv_speed = fb.pivot_table(values=['SprintSpeed'],
                           index=['Club'],
                           columns=['Position'],
                           aggfunc='mean',
                           fill_value=0)
piv_speed.info()
```
